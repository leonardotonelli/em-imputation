```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from data.real_example.preprocessing import load_data_binary
# Evaluate imputation methods on unanimous expert labels
from utils.real_example.evaluation import evaluate_imputers, visualize_evaluation_results

X, y_groundtruth, y_experts = load_data_binary(data_file='data\\real_example\\data.csv', gt_file='data\\real_example\\ground_truth.xlsx')

# 1. Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 2. Apply PCA (Retaining 95% variance)
pca = PCA(n_components=0.90)
X_pca = pca.fit_transform(X_scaled)

# Run a small experiment (reduce `n_repeats` for quicker runs)
results_df = evaluate_imputers(X_pca, y_experts, y_groundtruth)
agg = visualize_evaluation_results(results_df, save_dir='tests')

# Display a quick summary
results_df
```
